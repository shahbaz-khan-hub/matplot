```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
df=pd.read_csv("PK COVID-19-10may.csv")
df
pd.set_option("display.max_rows",None)
x=df["Deaths"]
y=df["Recovered"]
z=df["Cases"]
plt.plot(z,x, color="red",label="Deaths")
plt.plot(z,y, color="green",label="Recovered")
plt.legend()
plt.show()
```

```python
df.describe()#this fucntion can take many statistical functions 
```

```python
Matrix:sn.set(style="white")
corr=df.corr()
mask=np.zeros_like(corr,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True
f, ax= plt.subplots(figsize=(13,10))
cmap=sn.diverging_palette(220,10, as_cmap=True)
sn.heatmap(corr,mask=mask, cmap=cmap, vmax=3, center=0,square=True)
plt.savefig('ds2_corr.png' , bbox_inches='tight')
df.corr()
```
